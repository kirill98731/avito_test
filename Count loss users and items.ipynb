# С учетом категории

```python
final_cat = df[need_ind][['item_starttime_qr', 'item_category', 'item_id', 'round_scale']].groupby(['item_starttime_qr', 'item_category', 'round_scale']).count()
```

```python
final_cat.reset_index(inplace=True)
```

```python
first_final_1 = final_cat[final_cat['item_starttime_qr'] == '2015Q4']
```

```python
second_final_1 = final_cat[final_cat['item_starttime_qr'] == '2016Q1']
```

```python
df_final_cat = first_final_1.merge(second_final_1, left_on=['item_category','round_scale'], right_on=['item_category','round_scale'], how='left', suffixes=['_2015', '_2016'])
```

```python
df_final_cat['16/15'] = df_final_cat['item_id_2016']/df_final_cat['item_id_2015']
```

```python
df_final_cat.head()
```

```python
ind_top = (df_final_cat['round_scale'] != 1) & (df_final_cat['round_scale'] != 2)
```

```python
mean_final = df_final_cat[ind_top][['item_category','16/15']].groupby('item_category').mean()
```

```python
mean_final.reset_index(inplace=True)
```

```python
mean_final.columns = ['Категория', 'Средний прирост']
```

```python
mean_final
```

```python
df_final_cat = df_final_cat.merge(mean_final, left_on=['item_category'], right_on=['item_category'], how='left', suffixes=['_real', '_mean'])
```

```python
df_final_cat['prognos_item']=df_final_cat['item_id_2015']*df_final_cat['16/15_mean'].apply(lambda x: math.ceil(x))
```

```python
df_final_cat['diff'] = (df_final_cat['prognos_item'] - df_final_cat['item_id_2016']).apply(lambda x: max(0, math.ceil(x)))
```

```python
df_final_cat[(df_final_cat['round_scale'] == 1) | (df_final_cat['round_scale'] == 2 )][['round_scale', 'diff']].groupby('round_scale').sum()
```

```python
df_final_cat.columns = ['Квартал 2015', 'Категория', 'Усредненная лояльность', 'Количество объявлений 2015',
       'Квартал 2016', 'Количество объявлений 2016', 'Прирост 2016/2015', 'Средний рост',
       'Прогноз', 'Разница']
```
